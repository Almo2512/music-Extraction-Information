```python
import opensmile
import librosa
import soundfile as sf
import os
import json
import numpy as np
from pathlib import Path
```

```python
class NumpyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, (np.int_, np.intc, np.intp, np.int8,
                          np.int16, np.int32, np.int64, np.uint8,
                          np.uint16, np.uint32, np.uint64)):
            return int(obj)
        elif isinstance(obj, (np.float_, np.float16, np.float32, np.float64)):
            return float(obj)
        elif isinstance(obj, (np.ndarray,)):
            return obj.tolist()
        return json.JSONEncoder.default(self, obj)
```

```python
class AudioFeatureExtractor:
    def __init__(self, segment_duration=2.0):
        self.segment_duration = segment_duration
        self.smile = opensmile.Smile(
            feature_set=opensmile.FeatureSet.eGeMAPS,
            feature_level=opensmile.FeatureLevel.LowLevelDescriptors,
        )
    
    def load_audio(self, audio_path):

        audio_path = Path(audio_path)
        if audio_path.suffix.lower() not in ['.mp3', '.wav']:
            raise ValueError("File must be in MP3 or WAV format.")
        
        # Load audio with librosa
        y, sr = librosa.load(audio_path, sr=None)
        
        return y, sr
    
    def extract_features(self, audio_path):
   
        y, sr = self.load_audio(audio_path)
        
        segment_samples = int(sr * self.segment_duration)
        total_segments = len(y) // segment_samples
        
        temp_wav = 'temp_processing.wav'
        segments_features = []
        
        try:
            for i in range(0, len(y), segment_samples):
                segment = y[i:i + segment_samples]
                
                if len(segment) < segment_samples:
                    continue
                
                sf.write(temp_wav, segment, sr)
                
                # Extract features for segment
                features = self.smile.process_file(temp_wav)
                
                segment_stats = {
                    col: {
                        'mean': float(features[col].mean()), 
                        'std': float(features[col].std()),
                        'max': float(features[col].max()),
                        'min': float(features[col].min())
                    }
                    for col in features.columns
                }
                
                segment_stats['timestamp'] = {
                    'start': float(i / sr), 
                    'end': float((i + segment_samples) / sr)
                }
                
                segments_features.append(segment_stats)
                
               
                current_segment = i // segment_samples + 1
                print(f"Processing segment {current_segment}/{total_segments} "
                      f"({(current_segment/total_segments*100):.1f}%)", end='\r')
                
        finally:
          
            if os.path.exists(temp_wav):
                os.remove(temp_wav)
            print() 
        
        return segments_features
    
    def save_features(self, features, output_path):
     
        output_path = Path(output_path)
        output_path.parent.mkdir(parents=True, exist_ok=True)
        
        with output_path.open('w') as f:
            json.dump(features, f, cls=NumpyEncoder, indent=4)
```

```python
def main():
    extractor = AudioFeatureExtractor(segment_duration=2.0)
    audio_file = "Death Grips - Get Got.mp3"
    output_file = "features.json"
    
    try:
        
        print(f"Processing {audio_file}...")
        features = extractor.extract_features(audio_file)
        
        
        extractor.save_features(features, output_file)
        print(f"Features successfully extracted and saved to {output_file}")
        print(f"Processed {len(features)} segments")
        
    except Exception as e:
        print(f"Error during processing: {e}")

if __name__ == "__main__":
    main()
```

```python
import json
from pprint import pprint
import pandas as pd
import matplotlib.pyplot as plt

def display_features():
    json_path = "features.json"
    
    # Einfache Anzeige
    print("Basic JSON Content:")
    print("-----------------")
    with open(json_path, 'r') as f:
        data = json.load(f)
        pprint(data, indent=2)

def display_features_detailed():
    json_path = "features.json"
    
    # Load JSON data
    with open(json_path, 'r') as f:
        data = json.load(f)
    
    # Flatten data for DataFrame
    flattened_data = []
    
    for segment in data:
        segment_data = {
            'Timestamp': f"{segment['timestamp']['start']:.1f}s - {segment['timestamp']['end']:.1f}s"
        }
        
        for feature_name, feature_stats in segment.items():
            if feature_name != 'timestamp':
                for stat_name, value in feature_stats.items():
                    segment_data[f"{feature_name}_{stat_name}"] = value
                    
        flattened_data.append(segment_data)
    
    # Create DataFrame
    df = pd.DataFrame(flattened_data)
    
    # Display settings
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', None)
    
    print("\nDetailed Feature Analysis:")
    print("-----------------------")
    print(df)
    
    # Statistics
    print("\nSummary Statistics:")
    print(f"Total segments: {len(data)}")
    print(f"Total duration: {data[-1]['timestamp']['end']:.1f} seconds")
    print(f"Features per segment: {len(data[0]) - 1}")

def plot_features():
    json_path = "features.json"
    
    with open(json_path, 'r') as f:
        data = json.load(f)
    
    # Get list of features (excluding timestamp)
    features = [key for key in data[0].keys() if key != 'timestamp']
    
    # Plot each feature
    for feature in features:
        times = [(seg['timestamp']['start'] + seg['timestamp']['end'])/2 for seg in data]
        means = [seg[feature]['mean'] for seg in data]
        
        plt.figure(figsize=(12, 6))
        plt.plot(times, means)
        plt.title(f'{feature} over time')
        plt.xlabel('Time (seconds)')
        plt.ylabel(f'{feature} (mean)')
        plt.grid(True)
        plt.show()

if __name__ == "__main__":
    print("1. Basic Display")
    print("2. Detailed Analysis")
    print("3. Plot Features")
    choice = input("Choose display option (1-3): ")
    
    try:
        if choice == '1':
            display_features()
        elif choice == '2':
            display_features_detailed()
        elif choice == '3':
            plot_features()
        else:
            print("Invalid choice")
    except FileNotFoundError:
        print("Error: features.json file not found in the current directory")
    except json.JSONDecodeError:
        print("Error: Invalid JSON format in features.json")
    except Exception as e:
        print(f"An error occurred: {str(e)}")
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
import json
import colorsys

class AudioToColorMapper:
    def __init__(self):
        self.scaler = StandardScaler()
        self.model = MLPRegressor(
            hidden_layer_sizes=(64, 32),
            max_iter=1000,
            random_state=42
        )
        
    def prepare_features(self, json_path):
        # Load features
        with open(json_path, 'r') as f:
            data = json.load(f)
            
        features_list = []
        for segment in data:
            segment_features = []
            for feature_name, feature_stats in segment.items():
                if feature_name != 'timestamp':
                    # Use mean and std as features
                    segment_features.extend([
                        feature_stats['mean'],
                        feature_stats['std']
                    ])
            features_list.append(segment_features)
            
        return np.array(features_list)

    def map_features_to_color(self, features):
        normalized = self.scaler.transform(features)
        
        hsv_values = self.model.predict(normalized)
        
        # Convert HSV to RGB
        colors = []
        for hsv in hsv_values:
            # Ensure HSV values are in valid ranges
            h = np.clip(hsv[0], 0, 1)  # Hue: 0-1
            s = np.clip(hsv[1], 0, 1)  # Saturation: 0-1
            v = np.clip(hsv[2], 0, 1)  # Value: 0-1
            
            rgb = colorsys.hsv_to_rgb(h, s, v)
            colors.append(rgb)
            
        return np.array(colors)

    def visualize_colors(self, colors, features):
        import matplotlib.pyplot as plt
        
        plt.figure(figsize=(15, 2))
        for i, color in enumerate(colors):
            plt.axvspan(i, i+1, color=color)
        
        plt.axis('off')
        plt.show()

        feature_names = ['Loudness', 'Pitch', 'Energy']
        for i, feature in enumerate(features):
            feature_values = feature[:3]  # Take the first 3 features (mean of loudness, pitch, energy)
            print(f"Color {i+1}:")
            print(f"  {feature_names[0]}: {feature_values[0]:.2f}")
            print(f"  {feature_names[1]}: {feature_values[1]:.2f}")
            print(f"  {feature_names[2]}: {feature_values[2]:.2f}")
            print()

def train_and_visualize():
    mapper = AudioToColorMapper()
    
    features = mapper.prepare_features('features.json')
  
    np.random.seed(42)
    target_colors = np.random.rand(len(features), 3)
    
    mapper.scaler.fit(features)
    normalized_features = mapper.scaler.transform(features)
    mapper.model.fit(normalized_features, target_colors)
    
    predicted_colors = mapper.map_features_to_color(features)
    mapper.visualize_colors(predicted_colors, features)
    
    return mapper

def map_realtime_audio(mapper, new_features):
  
    colors = mapper.map_features_to_color(new_features)
    return colors

if __name__ == "__main__":
    mapper = train_and_visualize()
```
